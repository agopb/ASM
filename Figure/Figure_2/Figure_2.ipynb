```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['font.family'] = 'Arial'

names1 = [
    'ASM (Ours)',
    'Transformer',
    'ConvLSTM',
    'ResNet',
    'U-Net']

names2 = [
    'ASM',
    'Transformer',
    'ConvLSTM',
    'ResNet',
    'U-Net']

ACC_dict = {}
RMSE_dict = {}

for name1, name2 in zip(names1, names2):
    path = f'{name2}.csv'
    df = pd.read_csv(path)
    ACC_dict[name1] = df['ACC'].values
    RMSE_dict[name1] = df['RMSE'].values

fig, axes = plt.subplots(1, 2, figsize=(14,5)) 

ax = axes[0]
lines = []
for name in names1:
    linestyle = '-' if name == 'ASM (Ours)' else '--'
    line, = ax.plot(ACC_dict[name], label=name, linestyle=linestyle, linewidth=3)
    lines.append(line)

ax.set_xlabel('Lead time (days)', fontsize=16)
ax.set_ylabel('ACC', fontsize=16)
ax.set_title('ACC', fontsize=16)
ax.tick_params(axis='both', labelsize=16)
ax.grid(True)

ax = axes[1]
for i, name in enumerate(names1):
    linestyle = '-' if name == 'ASM (Ours)' else '--'
    ax.plot(RMSE_dict[name], label=name, linestyle=linestyle, linewidth=3)

ax.set_xlabel('Lead time (days)', fontsize=16)
ax.set_ylabel('RMSE', fontsize=16)
ax.set_title('RMSE', fontsize=16)
ax.tick_params(axis='both', labelsize=16)
ax.grid(True)

fig.subplots_adjust(bottom=0.3) 
fig.legend(lines, names1, loc='lower center', ncol=5, fontsize=16, frameon=False)

plt.savefig('figure_2.png', dpi = 300, bbox_inches='tight')
plt.show()
```
